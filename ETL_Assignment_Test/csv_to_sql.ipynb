```python
column_names = list(dataframe.columns.values)
```

```python
## Helpful function for determining data type from https://medium.com/@mgangrade7/create-sql-table-using-python-for-loading-data-from-pandas-dataframe-985281dde307
def getColumnDataType(dataTypes):
    columnDataTypes = []

    for c in dataTypes:
        if c == 'int64':
            columnDataTypes.append('int')
        elif c == 'float64':
            columnDataTypes.append('float')
        elif c == 'bool':
            columnDataTypes.append('boolean')
        else:
            columnDataTypes.append('varchar')
    return columnDataTypes

columnDataTypes = getColumnDataType(dataframe.dtypes)
        
```

```python
createTableSQL = 'CREATE TABLE BigData.dbo.RawBigDataTest ('

for data in range(len(columnDataTypes)):
    createTableSQL += '\n' + column_names[data] + ' ' + columnDataTypes[data] + ','

createTableSQL = createTableSQL[:-1] + ');'

print(createTableSQL)
```

```python
sql.execute(createTableSQL)
```

```python
sql.execute('''SELECT * FROM BigData.dbo.RawBigDataTest''')

for row in sql.fetchall():
    print (row)   

sql.close()
```
