```python
data_path = "C:\\Users\\colem\\Desktop\\Senior\\BigData\\GE_BigData_Project\\GE_Outcomes_V2\\Data\\Compliance\\GE_Compliance.csv"

import pandas as pd

dataframe = pd.DataFrame(pd.read_csv(data_path))

dataframe
```

```python
college_table = dataframe.pivot_table(index='college',values='Assignments')
college_table.plot(kind='bar', title='Average Number of Assignments per Course based on College')
```

```python
students_table = dataframe.pivot_table(index='college',values='Students')
students_table.plot(kind='bar', title='Average Number of Students per Course based on College')
```
